```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
c = 3e8 # speed of light in m/s
G = 6.67430e-11 # gravitational constant in m^3 kg^-1 s^-2
earth_mass = 5.972e24 # mass of Earth in kg
earth_radius = 6.371e6 # distance from the center of the Earth in meters (Earth's radius)

def special_relativity(proper_time, velocity):
    """
    Calculate time dilation due to Special Relativity
    """
    return proper_time / np.sqrt(1 - (velocity**2 / c**2))

def general_relativity(proper_time, mass, distance):
    """
    Calculate time dilation due to General Relativity
    """
    return proper_time / np.sqrt(1 - (2 * G * mass) / (distance * c**2))

def twin_paradox(age, velocity, turn_around_time):
    """
    Calculates the age of the traveling twin when the stationary twin reaches a certain age
    """
    return 2 * special_relativity(turn_around_time, velocity) + (age - 2 * turn_around_time)

# Special Relativity
velocity = np.linspace(0, 0.99*c, 100) # array of velocities from 0 to 99% speed of light
dilated_time_special = special_relativity(1, velocity)

plt.figure(figsize=(10,5))
plt.plot(velocity/c, dilated_time_special)
plt.title('Time Dilation due to Special Relativity')
plt.xlabel('Velocity (c)')
plt.ylabel('Dilated Time (s)')
plt.grid(True)
plt.show()
```

The x-axis represents the velocity of an object as a fraction of the speed of light (from 0 to 0.99c), and the y-axis represents the dilated time for 1 second of proper time. As seen from the plot, as the velocity of the object approaches the speed of light, time dilation increases dramatically. This indicates that the faster an object moves (compared to a stationary observer), the slower the time experienced by the object, as predicted by Special Relativity.

```python
# General Relativity
mass_blackhole = 1e30 # mass of a black hole in kg (around 5000 solar masses)
distance = np.linspace(1, 3e11, 100) # distances from the black hole from 1 meter to 3e11 meters
dilated_time_general = general_relativity(1, mass_blackhole, distance)

plt.figure(figsize=(10,5))
plt.plot(distance, dilated_time_general)
plt.title('Time Dilation due to General Relativity')
plt.xlabel('Distance from Black Hole (m)')
plt.ylabel('Dilated Time (s)')
plt.grid(True)
plt.show()
```

```python
# Twin Paradox
age = 10 # 10 years
turn_around_time = 1 # 1 year
twin_ages = twin_paradox(age, velocity, turn_around_time)

plt.figure(figsize=(10,5))
plt.plot(velocity/c, twin_ages)
plt.title('Twin Paradox')
plt.xlabel('Velocity (c)')
plt.ylabel('Age of Traveling Twin (years)')
plt.grid(True)
plt.show()
```

The x-axis represents the velocity of the traveling twin as a fraction of the speed of light, and the y-axis represents the age of the traveling twin when the stationary twin reaches 10 years old. The plot shows that the traveling twin ages less the faster they travel. This illustrates the paradox where the twin who traveled at high speed and then returned is younger than their identical twin who stayed stationary.
