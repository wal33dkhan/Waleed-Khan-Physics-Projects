Linear Pendulum equation with trapezoid method (part 4)

```python
import math
import numpy as np
import matplotlib.pyplot as plt

g,L,k,A,fi = 1,1.0,0.0,0,0.66667

# Force function of a Pendulum (Linear model)
def f(theta,w,t):
    force = -(g/L)*(theta)-k*w+A*math.cos(fi*t)
    return (force)

# Initial Conditions
# theta = 1.0,3.1,0.0 and w = 0.0,0.0,1.0 were all tested
theta = 1.0
w = 0.0
t = 0.0
dt = 0.01
npoints = 1000

# Trapezoid method
# Lists for storage
list_theta = [theta]
list_w = [w]
list_time = [t]
for nstep in range(1,npoints):
    k1a = dt * w
    k1b = dt * f(theta, w, t)
    k2a = dt * (w + k1b)
    k2b = dt * f(theta + k1a, w + k1b, t + dt)
    theta = theta + (k1a + k2a) / 2
    w = w + (k1b + k2b) / 2

    # evolve time step
    t = t + dt

    list_theta.append(theta)
    list_w.append(w)
    list_time.append(t)

# plotting instructions
plt.plot(list_time, list_theta)
plt.plot(list_time, list_w) #,'o',markevery=10)
plt.xlabel('Time (arb. units)')
plt.ylabel('w (orange), Theta (blue)')
plt.grid(True)
plt.show()
```

Non-Linear Pendulum with Trapezoid and Runge-Kutta Method (part 7)

```python
# This program plots omega and theta as a function of time for both the Range Kutta and trapezoid methods
import math
import numpy as np
import matplotlib.pyplot as plt

g,L,k,A,fi = 1,1.0,0.0,0,0.66667

# Force function of a Pendulum (Non-linear)
def f_nonlin(theta,w,t):
    force = -(g/L)*math.sin(theta)-k*w+A*math.cos(fi*t)
    return (force)

# Initial Conditions
# theta = 1.0,3.1,0.0 and w = 0.0,0.0,1.0 were all tested
theta = 3.0
w = 0.0
t = 0.0
dt = 0.01
npoints = 1000

# Trapezoid method
# Lists for storage
list_theta = [theta]
list_w = [w]
list_time = [t]
for nstep in range(1,npoints):
    k1a = dt * w
    k1b = dt * f_nonlin(theta, w, t)
    k2a = dt * (w + k1b)
    k2b = dt * f_nonlin(theta + k1a, w + k1b, t + dt)
    theta = theta + (k1a + k2a) / 2
    w = w + (k1b + k2b) / 2

    # evolve time step
    t = t + dt

    list_theta.append(theta)
    list_w.append(w)
    list_time.append(t)
# Runge-Kutta Method
# Lists for storage
list_theta2 = [theta]
list_w2 = [w]

for nstep in range(1,npoints):
    k1a = dt * w
    k1b = dt * f_nonlin(theta,w,t)
    k2a = dt * (w+k1b/2)
    k2b = dt * f_nonlin(theta+k1a/2,w+k1b/2,t+dt/2)
    k3a = dt * (w+k2b/2)
    k3b = dt * f_nonlin(theta+k2a/2,w+k2b/2,t+dt/2)
    k4a = dt * (w+k3b)
    k4b = dt * f_nonlin(theta+k3a,w+k3b,t+dt)
    theta = theta + ((k1a+2)*(k2a+2)*(k3a+k4a))/6
    w = w + ((k1b+2)*(k2b+2)*(k3b+k4b))/6

    # evolve time step
    t = t + dt

    list_theta2.append(theta)
    list_w2.append(w)

# plotting instructions
plt.plot(list_time, list_theta, label="Theta (trapezoid)")
plt.plot(list_time, list_theta2, label="Theta (Runge-Kutta)")
plt.legend(loc="upper left")
plt.xlim(0,10)
plt.xlabel('Time (arb. units)')
plt.ylabel('Omega Runge-Kutta (orange), Omega Trapezoid (blue)')
plt.grid(True)
plt.show()
```

Damped-Driven Nonlinear Pendulum Runge-Kutta Method (part 9)

```python
import math
import numpy as np
import matplotlib.pyplot as plt

g, L, k, A, fi = 1, 1.0, 0.5, 0.9, 0.66667


# Force function of a Pendulum (Non-linear)
def f_nonlin(theta, w, t):
    force = -(g / L) * math.sin(theta) - k * w + A * math.cos(fi * t)
    return force


# Initial Conditions
theta = 3.0
w = 0.0
t = 0.0
dt = 0.01
npoints = 5000

transient = 165500

# Runge-Kutta Method
# Lists for storage
list_theta = [theta]
list_w = [w]
list_time = [t]
for nstep in range(1, npoints):
    k1a = dt * w
    k1b = dt * f_nonlin(theta, w, t)
    k2a = dt * (w + k1b / 2)
    k2b = dt * f_nonlin(theta + k1a / 2, w + k1b / 2, t + dt / 2)
    k3a = dt * (w + k2b / 2)
    k3b = dt * f_nonlin(theta + k2a / 2, w + k2b / 2, t + dt / 2)
    k4a = dt * (w + k3b)
    k4b = dt * f_nonlin(theta + k3a, w + k3b, t + dt)
    theta = theta + (k1a + 2 * k2a + 2 * k3a + k4a) / 6
    w = w + (k1b + 2 * k2b + 2 * k3b + k4b) / 6

    # Restrict theta to lie within [-pi, pi]
    if theta > math.pi:
        theta = theta - 2 * math.pi
    elif theta < -math.pi:
        theta = theta + 2 * math.pi

    # evolve time step
    t = t + dt

    list_theta.append(theta)
    list_w.append(w)
    list_time.append(t)

# Parameter to plot only for t > t[transient]
transient_par = int(transient*dt)

plt.plot(list_time[transient_par:], list_theta[transient_par:])
plt.plot(list_time[transient_par:], list_w[transient_par:])
plt.xlabel('Time (arb. units)')
plt.ylabel('ω (orange), θ (blue)')
plt.grid(True)

plt.figure()
plt.scatter(list_theta[transient_par:], list_w[transient_par:],s=0.08)
plt.xlabel('θ (rad)')
plt.ylabel('ω (rad/s)')
plt.grid(True)
plt.show()
```
